```python
import pandas as pd
import glob
import matplotlib.pyplot as plt

path =r'C:\Users\slalonde\Documents\Personal\DataChallenge\output\USGS'
filenames = glob.glob(path + "\*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all csv data into one DataFrame
fullDf = pd.concat(dfs, ignore_index=True)
filterDf = fullDf[(fullDf['year'] == 2000) | (fullDf['year'] == 2010) | (fullDf['year'] == 2015) ].copy()
filterDf = filterDf[['year','Irrigation, Crop total self-supplied withdrawals for crops, fresh, in Mgal/d','county_nm']]
pivotDf = filterDf.pivot(index='year', columns='county_nm', values='Irrigation, Crop total self-supplied withdrawals for crops, fresh, in Mgal/d').astype(float)
pivotDf.loc[:,['Fresno County','Tulare County', 'Kern County', 'Lassen County', 'Monterey County']].plot()
```
